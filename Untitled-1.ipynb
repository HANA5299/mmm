```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import random as rand
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
```

```python
m = 2  
b = 3
x = np.random.randint(0,100,size=100)
y = m*x + b
noise = np.random.normal(0,1,size=len(x))
y_noise = y + noise
df = pd.DataFrame({ 'x':x,'y':y_noise})
df.head()
```

```python
x = df[['x']]
y = df['y']
scaled = StandardScaler()
scaled_x = scaled.fit_transform(x)
X_train , X_test , Y_train, Y_test = train_test_split(scaled_x,y,test_size=0.2,random_state=42)
x_list_train = X_train.tolist() 
print(x_list_train)
print(Y_test.head())
```

```python
new_model = LinearRegression()
model = new_model.fit(X_train,Y_train)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
testing_predictions = model.predict(X_test)
trainning_predictions = model.predict(X_train)
```

```python
mse_trainning = mean_squared_error(Y_train,trainning_predictions)
mse_testing = mean_squared_error(Y_test, testing_predictions)
mae_testing = mean_absolute_error(Y_test, testing_predictions)
mae_trainning = mean_absolute_error(Y_train, trainning_predictions)

r2 = r2_score(Y_test, testing_predictions)
r22 = r2_score(Y_train, trainning_predictions)

print("Mean Squared Error of trainning dataset ", mse_trainning)
print("Mean Squared Error of testing dataset ", mse_testing)

print("Mean Absolute Error of trainning dataset",mae_trainning )
print("Mean Absolute Error of testing dataset",mae_testing )

print("Variance Score of trainning data",r22)
print("Variance Score of testing data",r22)

```

based on error output 
the trainning and the testing errors are relatively low making the model mostfully ideal
this is because the model is linear with low amount of noise it may suffers from overfitting as the difference between training and testing error is not relatively 0


```python
plt.figure(figsize=(8,6))
plt.scatter(X_train, Y_train, color='blue', label='Train actual')
plt.scatter(X_train, trainning_predictions, color='pink', label='Train predicted')
plt.scatter(X_test, Y_test, color='red', label='Test actual')
plt.scatter(X_test, testing_predictions, color='purple', label='Test predicted')

plt.xlabel('X')
plt.ylabel('y')
plt.title('Training vs Testing Performance')
plt.legend()
plt.show()
```

```python
#formulating the multilinear equation
x1 = np.random.randint(0,100,size=100)
x2 = np.random.randint(0,100,size=100)
# a1 ,a2 and c are initialized as random values
a1 = rand.randint(0,10)
a2 = rand.randint(0,10)
c = rand.randint(0,20)
y = a1*x1 + a2*x2 + c
# adding noise
noise = np.random.normal(0,1,size=len(x1))
y = y + noise

dff = pd.DataFrame({
    'x1': x1,
    'x2': x2,
    'y': y
})
dff
```

```python
#splitting data 
new_x = dff[['x1','x2']]
new_y = dff['y']
x_train , x_test , y_train ,y_test = train_test_split(new_x,new_y,test_size=0.2,random_state=42)
x_train_scaled = scaled.fit_transform(x_train)
x_test_scaled = scaled.transform(x_test)
```

```python
multi_linear_model = LinearRegression()
multi = multi_linear_model.fit(x_train_scaled,y_train)
multi_test_prediction = multi.predict(x_test_scaled)
multi_train_prediction = multi.predict(x_train_scaled)
```

```python
multi_mse_trainning = mean_squared_error(y_train,multi_train_prediction)
multi_mse_testing = mean_squared_error(y_test, multi_test_prediction)
multi_mae_testing = mean_absolute_error(y_test, multi_test_prediction)
multi_mae_trainning = mean_absolute_error(y_train, multi_train_prediction)

multi_r2 = r2_score(y_test, multi_test_prediction)
multi_r22 = r2_score(y_train, multi_train_prediction)

print("Mean Squared Error of trainning dataset ", multi_mse_trainning)
print("Mean Squared Error of testing dataset ", multi_mse_testing)

print("Mean Absolute Error of trainning dataset",multi_mae_trainning )
print("Mean Absolute Error of testing dataset",multi_mae_testing )

print("Variance Score of trainning data",multi_r22)
print("Variance Score of testing data",multi_r22)
```
